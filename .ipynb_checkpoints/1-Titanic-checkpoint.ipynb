
### Parametros de la funcion read_csv

- sep = "," or sep = " " -> Separacion del archivo, comas, espacios, etc.
- dtype=None, dtype=({"pclass": np.float64, "edad": np.int32}) ->Sirve para saber que tipo de dato se usa para las columnas
- header=None, header=0 ->Cabecera para los nombres de las columas (que fila se usa como cabecera)
- names=None, names=["ingresos", "edad"] ->Sirve para identificar el nombre de las columnas en array
- skiprows=None skiprows=12 -> Se usa para saltar un determinado número de filas a la hora de leer el fichero o dataset
- index_col=None -> Sirve para poner identificador o etiqueta a la fila
- skip_blank_lines=False -> Las lineas en blanco se saltan o los NaN
- na_filter=False -> booleano que detecta los valores que faltan en el dataset (en blanco, NaN), pero se carga toda la fila

## Carga de datos a través de la función open

```python
d3 = open("../python-ml-course-master/datasets/customer-churn-model/Customer Churn Model.txt", "r")
```

```python
cols = d3.readline().strip().split(",")
n_cols = len(cols)
n_cols
```

```python
cols
```

Contar el número de filas

```python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
```

```python
for line in d3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1
    
print (f"El data set tiene {counter} y {n_cols} columnas")
```

```python
df3 = pd.DataFrame(main_dict)
df3.head()
```

## Lectura y escritura de ficheros

```python
infile = "../python-ml-course-master/datasets/customer-churn-model/Customer Churn Model.txt"
outfile = "../python-ml-course-master/datasets/customer-churn-model/Tab Customer Churn Model.txt"
```

```python
with open (infile, "r") as infile1:
    with open(outfile, "w") as outfile1:
        for line in infile1:
            fields = line.strip().split(",")
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
```

```python
df4 = pd.read_csv(outfile, sep = "\t")
```

```python
df4.head()
```

## Leer datos desde una URL

```python
medals_url = "http://winterolympicsmedals.com/medals.csv"
```

```python
medals_data = pd.read_csv(medals_url)
```

```python
medals_data.head()
```

```python
import csv
import urllib3
```

```python
http = urllib3.PoolManager()
r = http.request("GET", medals_url)
response = (r.data).decode("utf-8").split("\n")
response
```

```python
ruta = "./archivo_prueba.txt"
with open(ruta, "w") as archivo_nuevo:
        [archivo_nuevo.write(line + "\n") for line in response] # creamos nuestro txt
archivo_nuevo
```

```python
dataframe = open(ruta, "r")
columnas = dataframe.readline().strip().split(",") # obtenemos las columnas de la primera fila
main_dict = {columna:[] for columna in columnas} # creamos un diccionario con la key "columna" y la value [] (lista vacia)
[main_dict[columnas[i]].append(linea.strip().split(",")[i]) for linea in dataframe for i in range(len(columnas))] 

# crear dataframe
 
df_pd = pd.DataFrame(main_dict)
df_pd.head()
```
